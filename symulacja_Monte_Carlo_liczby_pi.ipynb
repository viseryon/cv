```python
from matplotlib import pyplot as plt
from matplotlib import style
import random
```

```python
n = int(input('podaj liczbę symulacji: '))
```

```python
inside = 0

x_ins = []
y_ins = []
x_out = []
y_out = []

pi = []

for nr in range(n+1):
    x = random.random()
    y = random.random()
    if x ** 2 + y ** 2 <= 1:
        inside += 1

        x_ins.append(x)
        y_ins.append(y)

    else:
        x_out.append(x)
        y_out.append(y)

    for i in range(6):
        if nr == 10 ** i:
            print()
            print(f'po 10^{i} symulacjach\npi równa się {inside/nr * 4}')

    pi.append(inside/(nr+1)*4)

style.use('dark_background')
fig, ax = plt.subplots(1, 2, figsize=(11, 5))
ax[0].plot(pi)
ax[0].set_xlabel('liczba symulacji')
ax[0].set_ylabel('przybliżenie liczby pi')
ax[0].set_title('przybliżenie liczby pi po x symulacjach')

ax[1].scatter(x_ins, y_ins)
ax[1].scatter(x_out, y_out, c='red')
ax[1].set_xlabel('wylosowane x')
ax[1].set_ylabel('wylosowane y')
ax[1].set_title(f'wylosowane (x, y) po {n} symulacjach')

fig.suptitle('Symulacja Monte Carlo liczby pi')
plt.show()
```
